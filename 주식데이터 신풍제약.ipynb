# 여기서 불러와서 사용하기

```python
# 바탕화면에서 불러오기
import pandas as pd
naver=pd.read_csv('C:/datapy/sinpoong.csv',parse_dates=["Date"])
naver
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
plt.plot(naver["Open"])
```

```python
plt.figure(figsize=(15, 10))
plt.plot(naver["Volume"])
```

```python
df_date =naver[["Open","Close","Date","Volume","Adj Close","High","Low","Change"]]
df_date
```

```python
df_date['quarter'], df_date['year'] = \
(df_date['Date'].dt.quarter, df_date['Date'].dt.year)
df_date.head()
```

```python
df_date['Date'].unique().shape
```

```python
df_date20 = df_date[df_date['year'] == '2020']
df_date20
```

```python
# 2020 년 데이터 추출

df2020 = df_date[df_date.year > 2019]
df2020
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df2020["Date"],df2020["Open"])
```

```python
#  날짜에 따른 거래량분석 신풍1년
plt.figure(figsize=(15, 10))
plt.plot(df2020["Date"],df2020["Volume"])
```

```python
# 2020 년 데이터 추출 3분기와 4분기 데이터 추출

df202034 = df2020[df_date.quarter > 2]
df202034
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df202034["Date"],df202034["Open"])
```

```python
#  날짜에 따른 거래량분석
plt.figure(figsize=(15, 10))
plt.plot(df202034["Date"],df202034["Volume"])
```

```python
# 거래량과 주식가격  
fig = plt.figure(figsize=(12, 8))
top_axes = plt.subplot2grid((4,4), (0,0), rowspan=3, colspan=4)
bottom_axes = plt.subplot2grid((4,4), (3,0), rowspan=1, colspan=4)
top_axes.plot(df202034["Date"],df202034["Open"],color='red')
top_axes.plot(df202034["Date"],df202034["Close"],color='blue')

bottom_axes.bar(df202034["Date"],df202034["Volume"])

plt.show()

```

```python
# 신풍제약 1년
opn = df20['Open']
high = df20['High']
low = df20['Low']
close = df20['Close']
volume = df20['Volume']
change = df20['Change']
x = df20['Date']
```

```python
from matplotlib import gridspec

plt.figure(figsize=(12, 5))                  # (12, 5) 크기의 판을 만든다.
gs = gridspec.GridSpec(nrows=2, ncols=1,     # 커다란 판을 위 아래 2개로 구획한다.
                       height_ratios=[5, 1]) # 각 구획의 높이 비율을 5:1로 한다.

##############################################
plt.subplot(gs[0])                           # 1번 구획을 불러들인다.
# 캔들스틱 그리는 코드 넣을 자리 


plt.subplot(gs[1])                           # 2번 구획을 불러들인다.
# 거래량 막대 그래프 그리는 코드 넣을 자리


##############################################

plt.subplots_adjust(wspace = 0, hspace = 0)  # 구획(subplot) 간의 간격을 없앤다.
plt.show()
```

```python
cmap = list(map(lambda x: 'red' if x >0 else 'c', change))

```

```python
from matplotlib import gridspec

plt.figure(figsize=(12, 5))                  # (12, 5) 크기의 판을 만든다.
gs = gridspec.GridSpec(nrows=2, ncols=1,     # 커다란 판을 위 아래 2개로 구획한다.
                       height_ratios=[5, 1]) # 각 구획의 높이 비율을 5:1로 한다.

##############################################
plt.subplot(gs[0])                           # 1번 구획을 불러들린다.
plt.title("신풍제약 주가", fontsize=15)
plt.bar(x, height=close-opn, bottom=opn, width=1, color=cmap)
plt.vlines(x, low, high, colors=cmap)
plt.xticks([])

plt.subplot(gs[1])                           # 2번 구획을 불러들린다.
plt.bar(x, volume, width=0.8, color=cmap)


##############################################

plt.subplots_adjust(wspace = 0, hspace = 0)  # 구획(subplot) 간의 간격을 없앤다.
plt.show()
```
